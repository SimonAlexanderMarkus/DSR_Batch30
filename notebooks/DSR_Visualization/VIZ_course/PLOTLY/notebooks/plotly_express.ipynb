```python
import plotly.express as px

gapminder = px.data.gapminder()
gapminder
```

```python
fig = px.scatter(
    data_frame=gapminder[gapminder.year.isin([1952, 2007])],
    x="gdpPercap",
    y="lifeExp",
    log_x=True,
    color="continent",
    size="pop",
    size_max=60,
    facet_col="year",
    width=800,
    height=500,
    title="Life Exp. vs. GDP",
)
```

```python
fig.show()
```

```python
fig = px.choropleth(
    data_frame=gapminder,
    locationmode="country names",
    locations="country",
    color="gdpPercap",
    animation_frame="year",
    width=800,
    height=600,
)
```

```python
fig.show()
```
